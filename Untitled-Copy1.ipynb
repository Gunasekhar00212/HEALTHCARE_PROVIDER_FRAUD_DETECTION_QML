```python
import pandas as pd

from qiskit_aer import Aer
from qiskit.circuit.library import TwoLocal, ZZFeatureMap
from qiskit_machine_learning.algorithms import VQC
from qiskit_machine_learning.optimizers import COBYLA
from qiskit.primitives import Sampler  # replaces QuantumInstance
from sklearn.metrics import accuracy_score, classification_report
from qiskit.primitives import StatevectorSampler
from qiskit_aer import Aer
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
beneficary_df=pd.read_csv('Train_Beneficiarydata-1542865627584.csv')
inpatient_df = pd.read_csv("Train_Inpatientdata-1542865627584.csv")
outpatient_df = pd.read_csv("Train_Outpatientdata-1542865627584.csv")
train_label_df = pd.read_csv("Train-1542865627584.csv")
```

```python
# Step 2: Label Inpatient / Outpatient
inpatient_df['Type'] = 'Inpatient'
outpatient_df['Type'] = 'Outpatient'
```

```python
claims_df = pd.concat([inpatient_df, outpatient_df], axis=0)
claims_df
```

```python
claims_with_bene = claims_df.merge(beneficary_df, on='BeneID', how='left')
```

```python
full_df = claims_with_bene.merge(train_label_df, on='Provider', how='left')
```

```python
full_df
```

```python
agg_features = claims_with_bene.groupby('Provider').agg({
    'InscClaimAmtReimbursed': ['sum', 'mean', 'max'],  # total/average/max claim money
    'ClaimID': 'count',                                # total number of claims
    'Type': lambda x: (x=='Inpatient').sum(),          # number of inpatient claims
    'ChronicCond_Diabetes': 'mean',
    'ChronicCond_IschemicHeart': 'mean',
    'ChronicCond_Osteoporasis': 'mean'
}).reset_index()

# Flatten column names
agg_features.columns = ['Provider', 'TotalReimbursed', 'AvgReimbursed', 'MaxReimbursed',
                        'TotalClaims', 'InpatientClaims', 'AvgDiabetes', 'AvgHeartDisease', 'AvgOsteoporosis']
```

```python
agg_features = agg_features.merge(train_label_df, on='Provider', how='left')
agg_features['FraudLabel'] = agg_features['PotentialFraud'].map({'Yes':1, 'No':0})
```

```python
X = agg_features.drop(columns=['Provider', 'PotentialFraud', 'FraudLabel'])  # features
y = agg_features['FraudLabel']                                              # target
X
```

```python
X = X.fillna(X.median())  # replace missing numbers with median
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=3)  # 3 features for 3 qubits
X_qm = pca.fit_transform(X_scaled)

print("QML-ready features shape:", X_qm.shape)
print("Labels shape:", y.shape)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_qm, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
provider_ids = agg_features['Provider']  # save for mapping predictions later
```

```python
smote = SMOTE(random_state=42, k_neighbors=5)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

```python
num_qubits = X_train_res.shape[1]
X_features = X_train_res  # QML-ready features
y_labels = y_train_res
```

```python
feature_map = ZZFeatureMap(feature_dimension=num_qubits, reps=1, entanglement='linear')
```

```python
ansatz = TwoLocal(num_qubits=num_qubits, rotation_blocks='ry', entanglement_blocks='cz', reps=1)
```

```python


sampler = StatevectorSampler()  # now explicitly using 1024 shots
```

```python
vqc = VQC(
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=COBYLA(maxiter=30),
    sampler=sampler
)
```

```python
vqc.fit(X_features,y_labels)
```
